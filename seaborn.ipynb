<h1>check corelation how relate  df.corr() only works for numeric if there are string then make numeric only True </h1>

```python
df.corr(method = 'pearson',
    min_periods = 1,
    numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only = True),annot=True)
plt.show()
```

<h1>Seaborn Heatmap help us to visulaize the correlation b/n data with color coded format</h1>

```python

sns.heatmap(df.corr(numeric_only = True),annot=True)
plt.rcParams['figure.figsize']=(20,7) # help us to increase size 
plt.show()
```

```python
# group by continent  just use if there are string and obj either replace dropna or use numeric_only=True
df.groupby('Continent').mean(numeric_only=True )
```

```python
sns.heatmap(df.groupby('Continent').mean(numeric_only=True ).corr(),annot=True)
plt.rcParams['figure.figsize']=(10,7) # help us to increase size 
plt.show()
```

```python
df2=df.groupby('Continent').mean(numeric_only=True ).sort_values( by="2022 Population",ascending=False)
```

```python
#official 
df2.plot()
```

<h2> Shafel the coloumn and row use transpose() method </h2>

```python
df3=df2.transpose()
df4=df3[df3.columns[2:5]]
df4.plot()
```
